### 1. Boxplot

```python
# Boxplot for revs per mile
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for the specified manufacturers
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
filtered_data = df[df['Manufacturer'].isin(selected_manufacturers)]

# Create boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=filtered_data)
plt.title('Revs per Mile for Selected Car Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile')
plt.show()
```

### 2. Histogram

```python
# Histogram of MPG
# Create histograms
plt.figure(figsize=(8, 6))
plt.hist(df['MPG.city'], bins=20, alpha=0.5, label='MPG City')
plt.hist(df['MPG.highway'], bins=20, alpha=0.5, label='MPG Highway')
plt.title('MPG Distribution in City vs Highway')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

From these bar charts, the cars in the city are genrally doing less miles per gallon (MPG) of around 0-25MP compared to the cars on highway. This may not necessaarily equate to efficiency because cars in the city stop at traffic lights too often and therefore end up doing less MPG.

In broad terms, driving on the highway typically offers better fuel efficiency compared to city driving. This is primarily due to factors such as maintaining a steady speed, reduced idling time, and the impact of aerodynamics. Highways allow for consistent speeds over longer stretches, minimizing fuel consumption compared to the frequent stops and starts common in city traffic. However, individual variables such as vehicle type, driving habits, and road conditions can influence fuel efficiency, so the general trend may not always apply in every situation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Cars93.csv')

# Extract the 'Turn.circle' and 'Wheelbase' columns
turn_circle_data = df['Turn.circle']
wheelbase_data = df['Wheelbase']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(wheelbase_data, turn_circle_data, alpha=0.5)
plt.title('Scatter Plot of Turn.circle vs. Wheelbase')
plt.xlabel('Wheelbase')
plt.ylabel('Turn.circle')
plt.grid(True)
plt.show()
```

### 4. Barplot

```python
# Barplot for mean horsepower
# Calculate mean horsepower for each car type
mean_hp_by_type = df.groupby('Type')['Horsepower'].mean()

# Create bar plot

# Define colors for each bar
colors = ['skyblue', 'lightgreen', 'salmon', 'orange', 'yellow', 'lightblue', 'lightcoral']

# Plot bar plot with different colors for each bar
plt.figure(figsize=(10, 6))
mean_hp_by_type.plot(kind='bar', color=colors)
plt.title('Mean Horsepower by Car Type', fontsize=16)
plt.xlabel('Car Type', fontsize=14)
plt.ylabel('Mean Horsepower', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# plt.figure(figsize=(8, 6))
# mean_hp_by_type.plot(kind='bar')
# plt.title('Mean Horsepower by Car Type')
# plt.xlabel('Car Type')
# plt.ylabel('Mean Horsepower')
# plt.xticks(rotation=45)
# plt.show()

```
